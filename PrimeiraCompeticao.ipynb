```
def FiltroNull(dados, fator):
  novos_dados = [index for index in dados.columns if dados[index].isnull().sum()/dados.shape[0] < (1-fator)]
  return novos_dados

def FiltroRepeat(dados, fator):
  novos_dados = [index for index in dados.columns if dados[index].value_counts(dropna=False, normalize=True).values[0] < (1-fator)]
  return novos_dados
# Deletando colunas com mais de 95% dados faltando
col_filtrado = FiltroNull(train, 0.95)
#col_filtrado = FiltroNull(train, 0.95)
#teste_filtrado = train[col_filtrado]
col_filtrado
```
